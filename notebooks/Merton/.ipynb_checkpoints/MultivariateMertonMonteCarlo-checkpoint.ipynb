# INFERENCE

```python
model_param = MertonJumpsPoissonCovariance.get_parameters()
inference_param = MertonJumpsPoissonCovariance.get_inference_parameters()

inference_param.update({"train_jumps_intensity": True})

inference_param.update({"train_jumps_arrival": True}) # J
inference_param.update({"train_jumps_size": True})  # H (Z in Handbook)

inference_param["train_jumps_mean"] = True
inference_param["train_jumps_covariance"] = True

inference_param.update({"train_expected_returns":True})
inference_param.update({"train_diffusion_covariance":True})
    
inference_param.update({"nmc": 2000,
                        "burning": 1000,
                        "metrics_logs": 50})

burning = inference_param.get("burning")
```

```python
# inference from simulated date
mjpc = MertonJumpsPoissonCovariance(None, data_loader, None, **model_param)
#monte_carlo_parameters = mjpc.initialize_inference(data_loader, None, **inference_parameters)
monte_carlo_parameters = mjpc.inference(data_loader, **inference_param)
```
